# Terza lezione di laboratorio

La prima parte della terza esercitazione consiste nella simulazione dell'andamento del prezzo di un asset al fine di dare una stima del prezzo di un'opzione Call e di un'opzione Put su tale asset.
Vengono forniti il prezzo iniziale dell'asset $S(0)=100$, il prezzo di esercizio (strike price) $K=100$, il tempo di consegna (delivery time) $T=1$, il tasso di interesse privo di rischio (drift) $\mu=0.1$ e la volatilità $\sigma=0.25$.
Si ipotizza che il prezzo dell'asset segua un moto browniano geometrico con i parametri sopra indicati.


Si considerano $G=100$ gruppi di simulazioni, e ciascun gruppo fornisce una stima del prezzo dell'asset $\{S_i\}_{i=1}^{G=100}$ al tempo $T$, del prezzo dell'opzione Call $\{C_i\}_{i=1}^{G=100}$ e dell'opzione Put $\{P_i\}_{i=1}^{G=100}$. Ogni gruppo è composto da $N=100$ simulazioni. Grazie all'ipotesi che il prezzo dell'asset si comporti come un moto browniano geometrico, la sua evoluzione al tempo $T$ nella j-esima simulazione dell'i-esimo gruppo si ricava quindi dalla relazione $$S_{ij}(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(T)\right]=$$ $$= S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma Z\sqrt{T}\right]$$ dove $W(T)$ è una variabile random estratta da una distribuzione normale con valor medio nullo e varianza $T$, ovvero $\mathcal{N}(0, t)$, e $Z$ è una variabile random estratta da $\mathcal{N}(0, 1)$. La stima di $S_i(T)$ per l'i-esimo gruppo è $S_i(T)=\frac{1}{N}\sum_{j=1}^{N=100} S_{ij}(T)$.
Si ottengono poi le stime per l'i-esimo gruppo dei prezzi delle opzioni $C_i=e^{-\mu T}\max[0, S_i(T)-K]$ e $P_i=e^{-\mu T}\max[0, K-S_i(T)]$.
Si calcolano infine le medie delle $G=100$ stime ottenute $C = \frac{1}{G} \sum_{i=1}^{G=100} C_i$ e $P = \frac{1}{G} \sum_{i=1}^{G=100} P_i$ e le loro deviazioni standard (deviazioni standard della media) $$\sigma_{Cm} \simeq \sqrt{\frac{1}{G-1} \left[ \frac{1}{G} \sum_{i=1}^{G=100} C^2_i - \left( \frac{1}{G} \sum_{i=1}^{G=100} C_i \right)^2 \right]} $$ $$\sigma_{Pm} \simeq \sqrt{\frac{1}{G-1} \left[ \frac{1}{G} \sum_{i=1}^{G=100} P^2_i - \left( \frac{1}{G} \sum_{i=1}^{G=100} P_i \right)^2 \right]} $$

Nei grafici sottostanti si rappresentano i valori $C$ e $P$ con relative barre di errore $\sigma_{Cm}$ e $\sigma_{Pm}$ all'aumentare del numero di gruppi di simulazioni presi in considerazione, e quindi del numero totale di simulazioni effettuate per ottenere tali stime. Sono stati confrontati con i valori ottenuti tramite il metodo analitico $C=14.975790778311286$ e $P=5.4595325819072364$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

plt.figure()
x, f, error = np.loadtxt("Call_1_passo.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, label='measured average value')
plt.hlines(14.975790778311286, 0, 10000, label='expected value')
plt.xlabel('#simulations')
plt.ylabel('Call-option price')
plt.grid(True)
plt.title('Call-option price, sampling after T')
plt.legend()
plt.show()

plt.figure()
x, f, error = np.loadtxt("Put_1_passo.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, label='measured average value')
plt.hlines(5.4595325819072364, 0, 10000, label='expected value')
plt.xlabel('#simulations')
plt.ylabel('Put-option price')
plt.grid(True)
plt.title('Put-option price, sampling after T')
plt.legend()
plt.show()
```

La seconda parte della terza esercitazione ha gli stessi obiettivi e le stesse ipotesi della prima parte. La struttura della simulazione è mantenuta identica con l'unica eccezione della modalità con cui si stima l'evoluzione del prezzo dell'asset.
Questa viene fatta dividendo l'intervallo $[0, T]$ con $T=1$ in $100$ sottointervalli di ampiezza $dt=\frac{T}{100}$ e definendo $t_k=kdt$ si ottiene il valore del prezzo dell'asset al variare del tempo dalla relazione ricorsiva
$$S_{ij}(t_{k+1}) = S_{ij}(t_k) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{k+1}-t_k)+\sigma Z_{k+1}\sqrt{t_{k+1}-t_k}\right]$$ dove $Z_k$ è una variabile random estratta da $\mathcal N (0, 1)$.
Partendo quindi da $S(0)$ si itera quindi fino a $S_{ij}(T)$.

I grafici sottostanti rappresentano le stesse grandezze dei grafici precedenti.
Anche il contenuto dei grafici è equivalente perché il processo stocastico in esame è un moto browniano geometrico, e non viene commesso quindi errore di discretizzazione.

```python
plt.figure()
x, f, error = np.loadtxt("Call_100_passi.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, label='measured average value')
plt.hlines(14.975790778311286, 0, 10000, label='expected value')
plt.xlabel('#simulations')
plt.ylabel('Call-option price')
plt.grid(True)
plt.title('Call-option price, discretised sampling of step dt')
plt.legend()
plt.show()

plt.figure()
x, f, error = np.loadtxt("Put_100_passi.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, label='measured average value')
plt.hlines(5.4595325819072364, 0, 10000, label='expected value')
plt.xlabel('#simulations')
plt.ylabel('Put-option price')
plt.grid(True)
plt.title('Put-option price, discretised sampling of step dt')
plt.legend()
plt.show()
```
