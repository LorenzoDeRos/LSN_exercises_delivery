# Sesta lezione di laboratorio

In questa esercitazione è stato utilizzato un codice di Molecular Dynamics che simula un sistema di particelle sfruttando l'algoritmo di Metropolis.

Il programma riceve come input i parametri macroscopici del sistema che si vuole simulare, e i dati tecnici della simulazione, ovvero il numero di steps Monte Carlo da effettuare in ogni blocco, il numero totale di blocchi della simulazione e un parametro della probabilità della prova di transizione. La simulazione del movimento delle particelle viene infatti effettuate tramite l'algoritmo di Metropolis, che propone ad ogni molecola del sistema lo spostamento in una nuova posizione, estratta con probabilità uniforme in un cubo centrato nella precedente posizione della molecola e con lato passato come parametro di input. La probabilità di accettazione del Metropolis viene calcolata come il rapporto tra il peso di Boltzmann delle proposta configurazione e il peso di Boltzmann della precedente configurazione. Vengono quindi fatte delle prove in cui si stabilisce il lato di questo cubo con lo scopo di ottenere un'accettazione media delle proposte dell'algoritmo di Metropolis pari a circa $50%$.

A questo punto le $N$ molecole vengono inizializzate nei punti di un reticolo fcc e viene lasciato evolvere il sistema con lo scopo di equilibrarlo e di ottenere una configurazione delle particelle molto probabile per i parametri macroscopici scelti.

A partire dalla configurazione ottenuta, si è quindi lasciato nuovamente evolvere il sistema e si sono raccolte le misure di energia potenziale per numero di particelle e di pressione del sistema. Queste misure sono state utilizzate per studiare l'autocorrelazione tra le stesse, e stabilire quindi quale fosse il numero minimo di steps Monte Carlo che ogni blocco dovrebbe contenere, in modo che le sue misure siano sufficientemente scorrelate.

Utilizzando le stesse misure si è inoltre calcolato l'andamento della deviazione standard della media utilizzando il blocking method in funzione del numero di steps presenti in ogni blocco. Questa funzione è crescente e satura ad un valore limite in corrispondenza di un certo numero di steps presenti in ogni blocco, che viene quindi scelto come input per la simulazione vera e propria, in modo da non sottostimare l'errore relativo alle grandezze misurate.

Infine è possibile fare la simulazione e ottenere misure dell'energia potenziale per ogni particella e pressione del sistema nel caso del gas argon per ciascuna delle tre fasi: liquido, solido e gas. Oltre a queste misure viene anche riportata per ciascuna fase l'andamento della densità di probabilità che due molecole si trovino ad una certa distanza. Questa funzione è stata confrontata con quella ottenuta grazie al codice di Molecular Dynamics già utilizzato che effettua la simulazione integrando le equazioni del moto delle singole molecole (quarta lezione di laboratorio).

## Liquido

La prima simulazione che è stata effettuata è quella di un liquido utilizzando il codice Monte Carlo NVT. Con lo scopo di equilibrare il sistema, inizialmente è stata fatta una simulazione di $1000$ steps Monte Carlo a partire da una configurazione fcc delle particelle. Nella figura sottostante è possibile vedere l'andamento dell'energia potenziale per ogni particella e della pressione del sistema durante questi primi steps di equilibrazione.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

equil_epot = np.loadtxt("epot_equilibration.liquid")

plt.figure()
plt.plot(range(equil_epot.size), equil_epot)
plt.title('Liquid, potential energy equilibration')
plt.xlabel('#steps')
plt.ylabel('epot/N')
plt.grid(True)

equil_pres = np.loadtxt("pres_equilibration.liquid")

plt.figure()
plt.plot(range(equil_pres.size), equil_pres)
plt.title('Liquid, pressure equilibration')
plt.xlabel('#steps')
plt.ylabel('pres')
plt.grid(True)

plt.show()
```

È stata quindi salvata la posizione delle particelle ottenuta dopo l'equilibrazione, e a partire da questa, è stata effettuata una simulazione, formata da $5 \cdot 10^5$ steps Monte Carlo. Nella figura sottostante sono riportate la totalità delle misure effettuate durante la simulazione in funzione del numero di steps Monte Carlo.

```python
epot = np.loadtxt("epot_allvalues.liquid")
nmeas=epot.size

plt.figure()
plt.plot(range(epot.size), epot)
plt.title('Liquid, all the potential energy measures')
plt.xlabel('#steps')
plt.ylabel('epot/N')
plt.grid(True)

pres = np.loadtxt("pres_allvalues.liquid")

plt.figure()
plt.plot(range(pres.size), pres)
plt.title('Liquid, all the pressure measures')
plt.xlabel('#steps')
plt.ylabel('pres')
plt.grid(True)

plt.show()
```

A questo punto è stata calcolata l'autocorrelazione delle misure sopra riportate in funzione del numero di steps Monte Carlo considerati. Per rendere questo calcolo più rapido è stato utilizzato un codice C++.
Nelle figure sottostanti è possibile vedere l'andamento dell'autocorrelazione, e come questa si annulli dopo circa $300$ steps, che si traduce nella condizione per la grandezza minima dei blocchi della simulazione $\text{nsteps} \gg 100$. Questi devono infatti essere composti da un numero di misure molto maggiore del numero di misure tra loro correlate.

```python
epot_corr = np.loadtxt("chi_epot.liquid")

first_measures=1000

epot_corr_first = np.zeros(first_measures)

for i in range(first_measures):
    epot_corr_first[i]=epot_corr[i]
    
plt.figure(figsize=(10, 6))
plt.plot(range(first_measures), epot_corr_first)
plt.title('Liquid, potential energy first measures autocorrelation')
plt.xlabel('#steps')
plt.ylabel('chi_epot')
plt.grid(True)

plt.figure(figsize=(10, 6))
plt.plot(range(epot_corr.size), epot_corr)
plt.title('Liquid, potential energy all the measures autocorrelation')
plt.xlabel('#steps')
plt.ylabel('chi_epot')
plt.grid(True)

pres_corr = np.loadtxt("chi_pres.liquid")

pres_corr_first = np.zeros(first_measures)

for i in range(first_measures):
    pres_corr_first[i]=pres_corr[i]
    
plt.figure(figsize=(10, 6))
plt.plot(range(first_measures), pres_corr_first)
plt.title('Liquid, pressure first measures autocorrelation')
plt.xlabel('#steps')
plt.ylabel('chi_pres')
plt.grid(True)

plt.figure(figsize=(10, 6))
plt.plot(range(pres_corr.size), pres_corr)
plt.title('Liquid, pressure all the measures autocorrelation')
plt.xlabel('#steps')
plt.ylabel('chi_pres')
plt.grid(True)

plt.show()
```

È stata quindi calcolato l'andamento della deviazione standard della media dei blocchi al variare del numero di steps Monte Carlo contenuti in ogni blocco. Come candidati a numero di steps per ogni blocco sono stati presi in considerazione tutti i divisori di $5 \cdot 10^5$ appartenenti all'intervallo $[10, 5000]$.
Tale andamento è stato riportato nelle figure sottostanti ed è evidente che la deviazione standard della media dopo circa $1000$ misure giunge a saturazione. Si ottiene quindi la condizione $\text{nsteps}=1000$ che è sufficientemente grande da soddisfare la condizione per la lunghezza minima $\text{nsteps} \gg 100$ dovuta alla correlazione delle misure.

```python
a=0
for i in range(10, 5001):
    if(nmeas%i==0):
        a+=1
range_nsteps=np.zeros(a)
b=0
for i in range(10, 5001):
    if(nmeas%i==0):
        range_nsteps[b]=int(i)
        b+=1
epot_error=np.zeros(range_nsteps.size)
pres_error=np.zeros(range_nsteps.size)
c=0
epot_sum2=0
pres_sum2=0
for nsteps in range_nsteps:
    nsteps=int(nsteps)
    nblocks=int(nmeas/nsteps)
    epot_ave=np.zeros(nblocks)
    pres_ave=np.zeros(nblocks)
    for iblock in range(nblocks):
        for istep in range(nsteps):
            epot_ave[iblock]+=epot[istep+iblock*nsteps]
            pres_ave[iblock]+=pres[istep+iblock*nsteps]
        epot_ave[iblock]=epot_ave[iblock]/nsteps
        pres_ave[iblock]=pres_ave[iblock]/nsteps
        epot_sum2+=epot_ave[iblock]*epot_ave[iblock]
        pres_sum2+=pres_ave[iblock]*pres_ave[iblock]
    epot_error[c]=math.sqrt((epot_sum2/nblocks-epot_ave.mean()*epot_ave.mean())/(nblocks-1))
    pres_error[c]=math.sqrt((pres_sum2/nblocks-pres_ave.mean()*pres_ave.mean())/(nblocks-1))
    epot_sum2=0
    pres_sum2=0
    c+=1

```

```python
plt.figure()
plt.plot(range_nsteps, epot_error)
plt.title('Liquid, potential energy standard deviation from the mean')
plt.xlabel('#steps in each block')
plt.ylabel('$\sigma_m$')
plt.grid(True)

plt.figure()
plt.plot(range_nsteps, pres_error)
plt.title('Liquid, pressure standard deviation from the mean')
plt.xlabel('#steps in each block')
plt.ylabel('$\sigma_m$')
plt.grid(True)

plt.show()
```

## Argon liquido

Con i dati ottenuti sopra è stata quindi effettuata la simulazione vera e propria di argon liquido. Di seguito sono riportati gli andamenti dell'energia potenziale per numero di particelle e della pressione del sistema in funzione del blocco considerato, e i valori medi progressivi di queste grandezze in funzione del numero di blocchi considerati.

```python
K_B=1.380649*10**(-23)
a_m_u=1.66053906660*10**(-27)
sigma=0.34*10**(-9)
epsilon=120*K_B
mass=39.948*a_m_u

fs_temp=epsilon/K_B
fs_step=sigma*math.sqrt(mass/epsilon)
fs_dist=sigma
fs_ene=epsilon
fs_pres=epsilon/sigma**3

x, epot_block, epot_ave, error_epot=np.loadtxt('argon.liquid.epot.0', usecols=(0, 1, 2, 3), unpack='True')
epot_block=epot_block*fs_ene*10**21
epot_ave=epot_ave*fs_ene*10**21
error_epot*error_epot*fs_ene*10**21

plt.figure(figsize=(10, 6))
plt.plot(x, epot_block)
plt.xlabel('#blocks')
plt.ylabel('epot [J x 10^-21]')
plt.grid(True)
plt.title('Liquid Argon, average potential energy in each block')

plt.figure(figsize=(10, 6))
plt.errorbar(x, epot_ave, error_epot)
plt.xlabel('#considered blocks')
plt.ylabel('epot [J x 10^-21]')
plt.grid(True)
plt.title('Liquid Argon, progressive average potential energy')

x, pres_block, pres_ave, error_pres=np.loadtxt('argon.liquid.pres.0', usecols=(0, 1, 2, 3), unpack='True')
pres_block=pres_block*fs_pres*10**(-7)
pres_ave=pres_ave*fs_pres*10**(-7)
error_pres=error_pres*fs_pres*10**(-7)

plt.figure(figsize=(10, 6))
plt.plot(x, pres_block)
plt.xlabel('#blocks')
plt.ylabel('pres [Pa x 10^7]')
plt.grid(True)
plt.title('Liquid Argon, average pressure in each block')

plt.figure(figsize=(10, 6))
plt.errorbar(x, pres_ave, error_pres)
plt.xlabel('#considered blocks')
plt.ylabel('pres [Pa x 10^7]')
plt.grid(True)
plt.title('Liquid Argon, progressive average pressure')

plt.show()
```

Viene inoltre riportata la denistà di probabilità di trovare due molecole ad una certa distanza in funzione di questa distanza. Questa è compatibile con quella ottenuta con il codice di Molecular Dynamics che effettua la simulazione integrando le equazioni del moto delle singole molecole.

```python
r, gdir, error_gdir = np.loadtxt("argon.liquid.gave.0", usecols=(0, 1, 2), delimiter=' ', unpack='True')
r=r*fs_dist*10**9

plt.figure(figsize=(10, 6))
plt.errorbar(r, gdir, error_gdir)
plt.xlabel('r [nm]')
plt.ylabel('g(r)')
plt.grid(True)
plt.title('Liquid Argon, distance probability distribution MonteCarlo')

r_c, gdir_c, error_gdir_c = np.loadtxt("gdir_ave_liquid.out", usecols=(0, 1, 2), delimiter=' ', unpack='True')
r_c=r_c*fs_dist*10**9

plt.figure(figsize=(10, 6))
plt.errorbar(r_c, gdir_c, error_gdir_c)
plt.xlabel('r [nm]')
plt.ylabel('g(r)')
plt.grid(True)
plt.title('Liquid Argon, distance probability distribution Verlet')

plt.show()
```

```python
x, epot_block, epot_ave, error_epot=np.loadtxt('argon.solid.epot.0', usecols=(0, 1, 2, 3), unpack='True')
epot_block=epot_block*fs_ene*10**21
epot_ave=epot_ave*fs_ene*10**21
error_epot*error_epot*fs_ene*10**21

plt.figure(figsize=(10, 6))
plt.plot(x, epot_block)
plt.xlabel('#blocks')
plt.ylabel('epot [J x 10^-21]')
plt.grid(True)
plt.title('Solid Argon, average potential energy in each block')

plt.figure(figsize=(10, 6))
plt.errorbar(x, epot_ave, error_epot)
plt.xlabel('#blocks')
plt.ylabel('epot [J x 10^-21]')
plt.grid(True)
plt.title('Solid Argon, progressive average potential energy')

x, pres_block, pres_ave, error_pres=np.loadtxt('argon.solid.pres.0', usecols=(0, 1, 2, 3), unpack='True')
pres_block=pres_block*fs_pres*10**(-7)
pres_ave=pres_ave*fs_pres*10**(-7)
error_pres=error_pres*fs_pres*10**(-7)

plt.figure(figsize=(10, 6))
plt.plot(x, pres_block)
plt.xlabel('#blocks')
plt.ylabel('pres [Pa x 10^7]')
plt.grid(True)
plt.title('Solid Argon, average pressure in each block')

plt.figure(figsize=(10, 6))
plt.errorbar(x, pres_ave, error_pres)
plt.xlabel('#blocks')
plt.ylabel('pres [Pa x 10^7]')
plt.grid(True)
plt.title('Solid Argon, progressive average pressure')

plt.show()
```
