## Esercizio 1

In questo esercizio è stato scritto un codice utilizzato l'algoritmo di Metropolis per il calcolo dell'energia del ground state dell'hamiltoniana monodimensionale
$$\hat{H}= -\frac{1}{2} \frac{d^2}{dx^2} + x^4 - \frac{5}{2}x^2$$
nelle unità $\hbar=1$ e $m=1$.
Per fare questo, poiché il ground state del potenziale $V(x)=x^2$ è una gaussiana e il potenziale sopra descritto nei suoi minimi presenta un comportamento asintoticamente analogo a quello parabolico, si è deciso di approssimare la funzione d'onda del ground state con una $\Psi_T$ costituita dalla somma di due gaussiane centrate in $\pm \mu$ e con deviazione standard $\sigma$.

A questo punto il valor medio dell'energia nello stato $\Psi_T$ può essere calcolato come segue
$$ \langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)} $$
dove nell'ultima uguaglianza è stata raccolta il modulo quadro di $\Psi_T$ normalizzata, che rappresenta quindi la densità di probabilità da campionare con l'algoritmo di Metropolis.

In questo codice la nuova posizione proposta dall'algoritmo viene estratta uniformemente all'interno di un segmento di lunghezza $L$. La probabilità di accettazione della nuova posizione proposta viene calcolata come il rapporto tra la densità di probabilità nella nuova posizione e la densità di probabilità nella posizione precedente.

## Esercizio 2

In questo esercizio è stato utilizzato il codice dell'esercizio precedente per ottimizzare i parametri di $\Psi_T$ in modo da minizzare l'energia di questo stato.
Questa ricerca è stata svolta manualmente, fino a giungere ai valori di $\mu=0.79$, $\sigma=0.62$ e $L=5.3$. Quest'ultimo parametro è stato scelto per fare in modo che l'accettazione dell'algoritmo fosse mediamente circa il $50\%$.
Il campionamento è costituito da $10^6$ steps Monte Carlo, suddivisi in $100$ blocchi da $10^4$ steps ciascuno.

È stato quindi calcolato il valor medio dell'energia in questo stato e la sua incertezza statistica utilizzando il blocking method. Di seguito viene riportato l'andamento del valor medio dell'energia all'aumentare del numero di blocchi considerati.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

measures, E_prog, error_prog=np.loadtxt("prog_E.txt", usecols=(0, 1, 2), delimiter=' ', unpack='True')

plt.figure(figsize=(10, 6))
plt.xlabel('#steps')
plt.ylabel('Energy')
plt.errorbar(measures, E_prog, error_prog)
plt.grid(True)
plt.title('Variational MonteCarlo, progressive average GS energy')

plt.show()
```

Di seguito viene inoltre riportata la densità di probabilità che è stata campionata dall'algoritmo di Metropolis, e il suo confronto con la densità di probabilità ottenuta dalla trasformazione dell'equazione di Schroedinger in un'equazione matriciale, e dalla sua diagonalizzazione.

```python
import numpy as np
import matplotlib.pyplot as plt

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])

# Plot a few things
plt.figure(figsize=(10,6))

n_bins=100
y = np.loadtxt("probability.my", usecols=(0), delimiter=' ', unpack='true')

f_my, bins, patches = plt.hist(y, n_bins, density=True, range=(-3,3))

x_my=np.zeros(f_my.size)
for i in range(x_my.size):
    x_my[i]=(bins[i]+bins[i+1])/2

plt.plot(x_my, f_my, label='$\Psi$ Variational MC')

scale = 0.3
plt.plot(x,(psi[0])**2, color='Green', label='$\Psi$ Matrix equation')
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.title("Potential and probability densities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.legend()
plt.show()
```

## Esercizio 8.3

In questa prima parte dell'esercizio è stato utilizzato il codice che implementa il metodo Path Integral Ground State per il campionamento del ground state di un hamiltoniana. Si procede quindi ad una descrizione dei concetti su cui l'algoritmo si basa.

È data un hamiltoniana $\hat{H}$ e identifichiamo con $|\Psi_n \rangle$ il suo insieme completo di autostati, a noi ignoti.
Supponiamo che il sistema sia monodimensionale e si trovi in un generico stato $|\Psi_T \rangle$. Questo stato evolve nel tempo come $|\Psi_T(t) \rangle=e^{-i\frac{t}{\hbar} \hat{H}} |\Psi_T \rangle$ e definendo $\tau:=i\frac{t}{\hbar}$ si può scrivere $|\Psi_T(\tau) \rangle=e^{-\tau\hat{H}} |\Psi_T \rangle$. Sviluppando poi $|\Psi_T \rangle$ sulla base di autostati si ottiene $|\Psi_T(\tau) \rangle = \sum_n c_n e^{-\tau E_n} |\Psi_n \rangle \sim c_0 e^{-\tau E_0}$. Questo significa che per tempi di propagazione sufficientemente lunghi, il termine dominante è la proiezione dello stato iniziale sul ground state, ovvero $|\Psi_0 \rangle = \lim\limits_{\tau \to \infty} e^{-\tau \hat{H}} |\Psi_T \rangle$. In rappresentazione di Schroedinger e introducendo $M$ volte la relazione di completezza è possibile scrivere
$$\Psi_0(x) = \lim\limits_{\tau \rightarrow \infty} \int dx_1 ... dx_M \langle x|e^{-\frac{\tau}{M} \hat{H}}|x_1 \rangle ... \langle x_{M-1} | e^{-\frac{\tau}{M}\hat{H}} | x_M \rangle \Psi_T(x_M) $$
Riconoscendo poi la definizione di propagatore $G\left(x_i, x_j, \frac{\tau}{M}\right):=\langle x_i | e^{-\frac{\tau}{M} \hat{H}} | x_j \rangle$ è possibile quindi calcolare
$$\Psi_0(x) \cong \int dx_1 ... dx_M G\left(x, x_1, \frac{\tau}{M}\right) ... G\left(x_{M-1}, x_M, \frac{\tau}{M}\right) \Psi_T (x_M)$$
Il valore di aspettazione di una qualunque osservabile $\hat{O}$ sul ground state può quindi essere approssimato con la quantità
$$\langle \hat{O} \rangle_0 = \frac{\int dx \Psi_0(x)^* \hat{O} \Psi_0(x)}{\langle \Psi_T(\tau) | \Psi_T(\tau) \rangle}
\cong \int \prod\limits_{i=1}^{2M+1} dx_i \hat{O}(x_{M+1}) \frac{\Psi_T(x_1) \prod\limits_{j=1}^{2M}G\left(x_j, x_{j+1}, \frac{\tau}{M}\right)\Psi_T(x_{2M+1})}{\langle \Psi_T(\tau) | \Psi_T(\tau) \rangle}$$
dove il secondo fattore dell'integrale costituisce una densità di probabilità da campionare.
I propagatori in generale e anche nel caso in esame non sono calcolabili analiticamente, per questo viene introdotta l'approssimazione primitiva
$$\langle x_i | e^{-\frac{\tau}{M} \hat{H}} | x_j \rangle \cong e^{-\frac{\tau}{2M} V(x_i)} \langle x_i | e^{-\frac{\tau}{M} \hat{T}} x_j \rangle e^{-\frac{\tau}{2M} V(x_j)}$$
che trascura termini $O\left(\left(\frac{\tau}{M}\right)^2\right)$ nell'esponenziale dell'operatore.
A questo punto il propagatore di un hamiltoniana puramente cinetica è analiticamente calcolabile
$$\langle x_i | e^{-\frac{\tau}{M} \hat{T}} | x_j \rangle = \frac{e^{-\frac{|x_i-x_j|^2}{4\gamma \tau / M}}}{(4\pi \gamma \tau /M)^{3/2}}$$
Questo propagatore ha inoltre la semplice interpretazione del peso di Boltzmann di un oscillatore armonico classico tra due punti in posizioni $x_i$ e $x_j$, che rende quindi possibile mappare una particella quantistica quale quella in esame su un polimero classico aperto formato da $2M+1$ particelle classiche interagenti tra loro con un potenziale armonico.

Sono state quindi utilizzato come $\Psi_T$ due funzioni: nel primo caso $\Psi_T=1$ e nel secondo caso $\Psi_T$ è stata scelta coincidere con la somma delle due gaussiane con parametri ottimizzati $\mu=0.79$ e $\sigma=0.62$ trovata nell'esercizio precedente.

Di seguito viene riportato il confrontro tra le densità di probabilità che sono state campionate dall'algoritmo Path Integral Groung State nel caso in cui $\Psi_T$ sia costante e nel caso in cui $\Psi_T$ sia la somma di gaussiane con parametri ottimizzati trovata nell'esercizio precedente. Vengono inoltre confrontate con la funzione d'onda ottenuta dall'equazione matriciale e la funzione d'onda ottimizzata ottenuta nell'esercizio precedente.
È possibile notare come utilizzando ciascuna delle 2 $\Psi_T$ si giunga ad una distribuzione di probabilità campionata molto più aderente a quella ottenuta utilizzando il metodo matriciale, rispetto invece a quella ottenuta con il metodo Monte Carlo variazionale. Questo significa che i propagatori dell'algoritmo PIGS hanno modificato la $\Psi_T$ fino a portarla a coincidere con l'effettivo ground state dell'hamiltoniana.

Si procede al'introduzione dei concetti su cui si basa il codice Path Integral Monte Carlo che è stato utilizzando nella seconda parte di questo esercizio.

Il valore di aspettazione di una qualunque osservabile $\hat{O}$ può essere calcolato come $\langle \hat{O} \rangle = \frac{\int dx \rho(x, x, \beta) O(x)}{Tr(\hat{\rho})}$ dove $\hat{\rho}(x, x', \beta)=\langle x | e^{-\beta\hat{H}} | x' \rangle $ è la matrice densità in rappresentazione di Schroedinger. Introducendo quindi $M$ volte la relazione di completezza all'interno dell'integrale è possibile ottenere
$$\langle \hat{O} \rangle = \int dx_1 ... dx_M O(x_1) \frac{\rho\left(x_1, x_2, \frac{\beta}{M}\right)...\rho\left(x_M, x_1, \frac{\beta}{M}\right)}{Tr(\hat{\rho})}$$
Il secondo fattore dell'integrale rappresenta quindi ancora una volta una distribuzione di probabilità che può essere campionata.
Acnora una volta gli operatori $\hat{\rho}$ non sono analiticamente calcolabili e viene nuovamente introdotta l'approssimazione primitiva
$$\rho\left(x_i, x_j, \frac{\beta}{M}\right)\cong e^{-\frac{\beta}{2M}V(x_i)} \frac{e^{-\frac{M|x_i-x_j|^2}{4\lambda \beta}}}{(4\pi \lambda \beta/M)^{3/2}} e^{-\frac{\beta}{2M}V(x_j)}$$
che permette ancora una volta di interpretare il sistema della singola particella quantistica come un polimero chiuso classico di $M$ particelle interagenti tra loro con un potenziale armonico. 

Di seguito vengono riportate le distribuzioni di probabilità che il codice PIMC campiona all'aumentare della temperatura. In particolare è possibile notare come a temperatura estremamente bassa il campionamente sia quasi esclusivo dello stato fondamentale, in quanto è sostanzialmente l'unico ad essere popolato. All'aumentare della temperatura, iniziano ad essere occupati anche stati ad energia maggiore, e quindi la distribuzione di probabilità che viene campionata è la sovrapposizione di tutti questi stati, che iniziano ad avere peso di Boltzmann sempre più simile tra loro. Il limite di questa procedure è un'onda piana all'interno della zona classicamente accessibile con decadimento esponenziale nelle zone classicamente inaccessibili del potenziale.
