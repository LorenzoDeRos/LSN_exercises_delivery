Andamento dell'energia in funzione del numero di blocchi considerati, una volta che i parametri del trial ground state sono stati ottimizzati.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

measures, E_prog, error_prog=np.loadtxt("prog_E.txt", usecols=(0, 1, 2), delimiter=' ', unpack='True')

plt.figure()
plt.xlabel('#steps')
plt.ylabel('Energy')
plt.errorbar(measures, E_prog, error_prog)
plt.grid(True)
plt.show()
```

```python
n_bins=100
f = np.loadtxt("function_plot.txt", usecols=(0), delimiter=' ', unpack='true')

n, bins, patches = plt.hist(f, n_bins, density=True, range=(-3,3))
print(bins.size)
print(n.size)
x=np.zeros(n.size)
for i in range(x.size):
    x[i]=(bins[i]+bins[i+1])/2

plt.figure()
plt.plot(x, n)

plt.xlabel('x')
plt.ylabel('psi_0')
plt.title('Histogram of the ground state')
plt.grid(True)
```
