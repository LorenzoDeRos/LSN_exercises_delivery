```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, err_f = np.loadtxt('probability.dat', usecols=(0, 1, 2), delimiter=' ', unpack='True')

plt.figure()
plt.errorbar(x, f, err_f)
plt.grid(True)
plt.show()
```

```python
b, T, err_T = np.loadtxt('kinetic.dat', usecols=(0, 1, 2), delimiter=' ', unpack='True')
c, V, err_V = np.loadtxt('potential.dat', usecols=(0, 1, 2), delimiter=' ', unpack='True')

plt.figure()
plt.errorbar(b, T, err_T)
plt.errorbar(c, V, err_V)
plt.grid(True)
plt.show()
```
