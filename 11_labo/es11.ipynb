## Esercizio 1

In questo esercizio viene migliorato il risultato del fit di alcuni punti ditribuiti lungo la retta $y=2x+1$, alla quale viene sommato del rumore gaussiano.

I parametri iniziale erano $\text{Ntrain}= 100$, $\text{Nvalid}= 10$, e $\text{Nepoch}= 30$, che rappresentano rispettivamente il numero di punti utilizzati per l'apprendimento della rete neurale, il numero di punti utilizzati come validazione per la rete neurale e il numero di epoche di durata dell'apprendimento.

Si è scelto di partire da $\sigma = 0.2$ ed è stato sufficiente portare $\text{Ntrain}= 1000$ e $\text{Nvalid}= 100$ mantenendo lo stesso $\text{Nepoch}=30$ per ottenere una stima del coefficiente angolare  e del termine noto con un errore percentuale che ha come ordine di grandezza lo $0.1\%$. Questo significa che il campione di dati prima fornito alla rete neurale non era sufficiente al suo apprendimento. Si è però notato anche che mantenendo invece $\text{Ntrain}= 100$ e $\text{Nvalid}=10$ ma aumentando $\text{Nepoch} = 200$ si giunge ad una stima del coefficiente angolare e del termine noto con un errore percentuale che ha come ordine di grandezza l'$1\%$. Ciò significa che con una scarsa quantità di dati, alla rete neurale occorre una maggiore quantità di epoche per apprendere, correndo però il rischio di overfitting dei dati.

Aumentando molto $\sigma$ è necessario anche aumentare $\text{Ntrain}$ e $\text{Nvalid}$ perché il trend rettilineo dei dati si confonde con il rumore crescente.

Di seguito viene riportato il fit lineare ottenuto aumentando il numero di punti utilizzati per train e validation.

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

### Aggiungere dati di train e validation

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

sigma = 0.2 # noise standard deviation, for the moment it is absent

Ntrain=1000
Nvalid=100
Npredict=100
Nepoch=30
```

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, Ntrain)
x_valid = np.random.uniform(-1, 1, Nvalid)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
plt.figure()
plt.plot(x_valid, y_target, label='target points')
plt.scatter(x_valid, y_valid, color='r', label='validation points')
plt.legend();
plt.grid(True);
plt.show()
```

```python
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

model.summary()
```

```python
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=Nepoch,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
model.get_weights()
```

```python
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
plt.figure()
x_predicted = np.random.uniform(-1, 1, Npredict)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r', label='predicted points')
plt.plot(x_valid, y_target, label='target points')
plt.grid(True)
plt.legend()
plt.show()
```

### Scegliere la struttura della rete neurale

Di seguito vengono riportate 3 fit ottenuti su reti neurali con circa pari numero di parametri, ma con una struttura della rete differente. In particolare la prima rete possiede una struttura del tipo (32, 7, 1) (numero di neuroni in ogni layer) con numero totale di parametri 303, la seconda rete del tipo (8, 8, 8, 8, 8, 1) e con numero di parametri 313 e la terza rete del tipo (1, 6, 21, 6, 1) con numero di parametri 300. Come si evince dai grafici sotto riportati, non sono visibili sostanziali differenze, nelle medesime situazioni di train, validation e numero di epoche, in quanto il mse finale è dello stesso ordine di grandezza. Questo ovviamente tenendo conto delle usuali oscillazioni di bontà che i fit presentano tra un funzionamento ed un altro. L'unica differenza percepibile risultano essere le oscillazioni in validation loss. Infatti la prima struttura sembra essere quella con la maggiore monotonia nell'andamento della validation loss, mentre l'ultima presenta le oscillazioni più importanti. Inoltre rispetto alla prima struttura, le altre due sembrano convergere più velocemente all'ordine di grandezza cercato per la validation loss, ma poi rallentano la convergenza e necessitano di un maggior numero di epoche per giungere ad un fit accettabile.

```python
def my_f(x):
    return 4-3*x-2*x*x+3*x*x*x

sigma = 0.2 # noise standard deviation, for the moment it is absent

Ntrain=10000
Nvalid=1000
Npredict=1000
Nepoch=70
```

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, Ntrain)
x_valid = np.random.uniform(-1, 1, Nvalid)
x_valid.sort()
y_target = my_f(x_valid) # ideal (target) parabolic function

y_train = np.random.normal(my_f(x_train), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(my_f(x_valid), sigma)
```

```python
plt.plot(x_valid, y_target, label='function to fit')
plt.scatter(x_train, y_train, color='g', label='train points')
plt.scatter(x_valid, y_valid, color='r', label='validation points')
plt.legend()

plt.grid(True); plt.show()
```

## Esercizio 3

In questo esercizio veniva richiesto di effettuare un fit della superficie $z=\sin(x^2+y^2)$ nel dominio $(x, y) \in \left[-\frac{3}{2}, \frac{3}{2}\right] \times \left[-\frac{3}{2}, \frac{3}{2}\right]$.
Per fare questo si è deciso di aumentare il punto di punti utilizzati per train a $10^4$ e validation a $10^3$, in modo da velocizzare l'apprendimento della rete. Alla superficie è stato aggiunto del rumore gaussiano con deviazione standard $0.1$. La rete è stata costruita utilizzando la struttura delle rete risultata più efficace nell'esercizio precedente, ovvero con un numero decrescente di nodi dall'input all'output, e in particolare della forma $(100, 75, 30, 1)$. La funzione di attivazione utilizzata è stata RELU nel layer di input e nei layer nascosti. Per l'output invece si è optato per la funzione di attivazione TANH in quanto questa funzione e la superficie da fittare hanno in comune il codominio $[-1, 1]$. Questo garantisce che la rete non possa prevedere per la superficie dei valori all'esterno di questo codominio. La quantità di nodi scelti con al durata di 70 epoche sono risultati sufficienti ad ottenere un fit della funzione senza rischi di overfitting.
Come ottimizzatore è stato scelto SGD e come loss MSE poiché sono risultati essere tra i più efficienti nell'esercizio precedente.

```python
Ntrain = 10000
Nvalid = 1000
Npredicted = 1000
Nepoch=70
sigma = 0.1

np.random.seed(0)

# generate synthetic training dataset
xy_train = np.random.uniform(-1.5, 1.5, (Ntrain, 2))
R2_train = np.zeros(Ntrain)
z_train = np.zeros(Ntrain)
for i in range(Ntrain):
    R2_train[i] = xy_train[i, 0]**2 + xy_train[i, 1]**2
    z_train[i] = np.sin(R2_train[i]) + np.random.normal(0, sigma)

# generate synthetic validation data
xy_valid = np.random.uniform(-1.5, 1.5, (Nvalid, 2))
R2_valid = np.zeros(Nvalid)
z_valid = np.zeros(Nvalid)
z_target = np.zeros(Nvalid)
for i in range(Nvalid):
    R2_valid[i] = xy_valid[i, 0]**2 + xy_valid[i, 1]**2
    z_valid[i] = np.sin(R2_valid[i]) + np.random.normal(0, sigma)
    z_target[i] = np.sin(R2_valid[i])
    
plt.scatter(np.sqrt(R2_train), z_train, label='train points')
plt.scatter(np.sqrt(R2_valid), z_valid, label='validation points')
plt.scatter(np.sqrt(R2_valid), z_target, label='target points')
plt.xlabel('r')
plt.grid(True)
plt.legend()

plt.show()
```

```python
model = Sequential()
model.add(Dense(100, input_shape=(2,), activation='relu'))
model.add(Dense(75, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(1, activation='tanh'))

model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

model.summary()
```

```python
history = model.fit(x=xy_train, y=z_train, 
          batch_size=32, epochs=Nepoch,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(xy_valid, z_valid))
```

```python
#model.get_weights()
```

```python
score = model.evaluate(xy_valid, z_valid, batch_size=32, verbose=1)
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
xy_predicted = np.random.uniform(-1.5, 1.5, (Npredicted, 2))
R2_predicted = np.zeros(Npredicted)
for i in range(Npredicted):
    R2_predicted[i] = xy_predicted[i, 0]**2 + xy_predicted[i, 1]**2
z_predicted = model.predict(xy_predicted)
plt.scatter(np.sqrt(R2_valid), z_target, label='target points')
plt.scatter(np.sqrt(R2_predicted), z_predicted,color='r', label='predicted points')
plt.xlabel('r')
plt.legend()
plt.grid(True)
plt.show()
```
