## Esercizio 11.1

In questo esercizio viene migliorato il risultato del fit di alcuni punti ditribuiti lungo la retta $y=2x+1$, alla quale viene sommato del rumore gaussiano.

I parametri iniziale erano $\text{Ntrain}= 100$, $\text{Nvalid}= 10$, e $\text{Nepoch}= 30$, che rappresentano rispettivamente il numero di punti utilizzati per l'apprendimento della rete neurale, il numero di punti utilizzati come validazione per la rete neurale e il numero di epoche di durata dell'apprendimento.

Si è scelto di partire da $\sigma = 0.2$ ed è stato sufficiente portare $\text{Ntrain}= 1000$ e $\text{Nvalid}= 100$ mantenendo lo stesso $\text{Nepoch}=30$ per ottenere una stima del coefficiente angolare  e del termine noto con un errore percentuale che ha come ordine di grandezza lo $0.1\%$. Questo significa che il campione di dati che il campione prima fornito alla rete neurale non era sufficiente al suo apprendimento. Si è però notato anche che mantenendo invece $\text{Ntrain}= 100$ e $\text{Nvalid}=10$ ma aumentando $\text{Nepoch} = 200$ si giunge ad una stima del coefficiente angolare e del termine noto con un errore percentuale che ha come ordine di grandezza l'$1\%$. Ciò significa che con una scarsa quantità di dati, alla rete neurale occorre una maggiore quantità di epoche per apprendere, correndo però il rischio di overfitting dei dati.

Aumentando molto $\sigma$ è necessario anche aumentare $\text{Ntrain}$ e $\text{Nvalid}$ perché il trend rettilineo dei dati si confonde con il rumore crescente.

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

sigma = 0.2 # noise standard deviation, for the moment it is absent

Ntrain=1000
Nvalid=100
Npredict=100
Nepoch=30
```

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, Ntrain)
x_valid = np.random.uniform(-1, 1, Nvalid)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

```python
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
model.summary()
```

```python
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=Nepoch,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
model.get_weights()
```

```python
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, Npredict)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

# RELU

```python
def my_f(x):
    return 4-3*x-2*x*x+3*x*x*x

sigma = 0.1 # noise standard deviation, for the moment it is absent

Ntrain=10000
Nvalid=1000
Npredict=1000
Nepoch=30
```

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, Ntrain)
x_valid = np.random.uniform(-1, 1, Nvalid)
x_valid.sort()
y_target = my_f(x_valid) # ideal (target) parabolic function

y_train = np.random.normal(my_f(x_train), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(my_f(x_valid), sigma)
```

```python
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

```python
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,), activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1, activation='relu'))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
model.summary()
```

```python
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=Nepoch,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
model.get_weights()
```

```python
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, Npredict)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, Npredict)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```
