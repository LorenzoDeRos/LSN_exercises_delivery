## Esercizi 1 e 2

Il codice di Molecular Dynamics è stato modificato in modo da poter funzionare in 3 modalità.
La scelta tra le 3 modalità viene effettuata da riga di comando, quando richiesto dal programma, e sono stati aggiunti al makefile dei comandi con lo scopo di rendere più rapido il processo di importazione degli input e di salvataggio degli output.

La prima modalità fa partire la simulazione da una configurazione di reticolo fcc contenuta in "config.0" e assegna velocità casuali alle particelle. Riscala poi tutte le velocità in modo da ottenere l'energia cinetica totale, e quindi la temperatura desiderata. Per fare questo sfrutta la relazione $<v>=\sqrt{3T}$ che lega il valor medio del modulo della velocità delle particelle alla temperatura.

La seconda modalità fa partire la simulazione da dove era stata lasciata, ovvero dalla configurazione letta da "config.final". Per assegnare le velocità alle particelle il programma utilizza anche il file "config_old.final", dove è stata scritta la configurazione del gas al time-step precedente all'ultimo time-step della simulazione precedente. In questa modalità all'inizio della simulazione viene calcolata l'energia cinetica totale del gas e confrontata con quella richiesta in input, per poi riscalare le velocità sfruttando la relazione già citata. Questo si rende necessario poiché quando il sistema parte da una configurazione fcc, tutte le particelle si trovano in minimi del potenziale, e hanno quindi mediamente energia cinetica superiore a quella attesa all'equilibrio. Si ripete la correzione del sistema finché le particelle non vengono a trovarsi in una configurazione di posizione e velocità probabile per la temperatura che è a loro richiesta.

La terza modalità, come la seconda, legge l'ultima configurazione da "config.final" e la precedente da "config_old.final", assegnando così posizioni e velocità, ma non riscala queste ultime, in quanto si assume che siano già state riscalate fino ad ottenere la temperatura desiderata.

L'evoluzione temporale del sistema è implementata da un algoritmo che integra l'equazione del moto delle singole particelle approssimandola fino al secondo ordine come segue
$$\vec{r}(t+dt)\approx2\vec{r}(t)−\vec{r}(t−dt)+\frac{dt^2}{m}\vec{F}(t)$$
Questa scelta ha l'importante caratteristica di essere invariante sotto inversione temporale, e questo ha come conseguenza un'ottima conservazione dell'energia del sistema. Questo è evidente nei grafici contenenti contemporaneamente energia cinetica, potenziale e totale. Infatti le energie cinetica e potenziale presentano delle oscillazioni a breve termine, ma la loro somma (energia totale) presenta delle oscillazioni molto minori.

Il potenziale di interazione delle singole particelle è stato supposto di tipo Lennard-Jones, ma con un raggio di cut-off oltre il quale l'interazione tra le due particelle viene trascurata.

È stata inoltre implementata nel codice fornito il calcolo e la stampa delle medie progressive con relativi errori in funzione del numero di blocchi di misure delle 4 grandezze: temperatura, energia potenziale, energia cinetica ed energia totale. Le energie da questo momento in poi saranno intese per divise per il numero di particelle del sistema.

### Simulazione di un liquido

Viene ora descritta la procedura per la simulazione di un generico liquido.

Per prima cosa viene effettuata la simulazione nella prima modalità, quindi partendo da fcc ("config.0") per $10^4$ step (misurando il valore della temperatura una volta ogni 10 steps in modo da ottenere $10^3$ misure). Come si vede dalla figura sottostante, il sistema non sta oscillando in un intorno della temperatura desiderata, perciò si fa ripartire la simulazione nella seconda modalita, e si ripete questo per un totale di 7 volte correggendo le velocità. Viene accettata come ragionevole la configurazione ottenuta dopo 7 riequilibrazioni e viene salvata la configurazione delle particelle, in modo da poter essere recuperata in un secondo momento.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

f = np.loadtxt("correction_liquid.dat")
x=np.zeros(f.size)
for i in range(x.size):
    x[i]=i+1
plt.figure()
plt.plot(x, f, label='measured value')
plt.hlines(1.9, 0, x.size, label='target value')
plt.xlabel('#measures')
plt.ylabel('temperature')
plt.grid(True)
plt.title('Liquid equilibration')
plt.legend()

plt.show()
```

Viene ora effettuata la simulazione vera e propria nel caso di un liquido. Per iniziare la simulazione dal liquido alla temperatura desiderata è sufficiente importare i parametri di input e la configurazioni finali salvate dopo il punto precedente. La simulazione viene quindi effettuata nella terza modalità, è composta $10^5$ steps e ottiene una misura delle grandezze ogni 10 steps, per un totale di $10^4$ misure.
Le misure effettuate durante la simulazione sono state salvate.
Vengono inoltre salvate anche le medie progressive delle grandezze e i relativi errori all'aumentare del numero di blocchi di misure considerati.

```python
plt.figure()
f1 = np.loadtxt("liquid_measures_epot.dat")
x=np.zeros(f1.size)
for i in range(x.size):
    x[i]=i+1
f2 = np.loadtxt("liquid_measures_ekin.dat")
f3 = np.loadtxt("liquid_measures_etot.dat")
f4 = np.loadtxt("liquid_measures_temp.dat")
plt.plot(x,f1, label='potential energy')
plt.plot(x,f2, label='kinetic energy')
plt.plot(x,f3, label='total energy')
plt.plot(x,f4, label='temperature')
plt.xlabel('#measures')
plt.ylabel('quantities')
plt.grid(True)
plt.title('Liquid, all the measures of the simulation')
plt.legend()

plt.show()
```

```python
plt.figure()
f1, error1 = np.loadtxt("liquid_ave_epot.dat", usecols=(0,1), delimiter=' ', unpack='true')
x=np.zeros(f1.size)
for i in range(x.size):
    x[i]=i+1
plt.errorbar(x,f1,error1)
plt.xlabel('#blocks')
plt.ylabel('epot')
plt.grid(True)
plt.title('Liquid, measured average potential energy')

plt.figure()
f2, error2 = np.loadtxt("liquid_ave_ekin.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,f2,error2)
plt.xlabel('#blocks')
plt.ylabel('ekin')
plt.grid(True)
plt.title('Liquid, measured average kinetic energy')

plt.figure()
f3, error3 = np.loadtxt("liquid_ave_etot.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,f3,error3)
plt.xlabel('#blocks')
plt.ylabel('etot')
plt.grid(True)
plt.title('Liquid, measured average total energy')

plt.figure()
f4, error4 = np.loadtxt("liquid_ave_temp.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,f4,error4, label='measured value')
plt.hlines(1.9, 0, x.size, label='target value')
plt.xlabel('#blocks')
plt.ylabel('temperature')
plt.grid(True)
plt.title('Liquid, measured average temperature')
plt.legend()

plt.show()
```

### Argon solido

La procedura sopra descritta viene ripetuta cambiando i parametri iniziali con quelli relativi ad argon solido.

Per prima cosa viene effettuata la simulazione nella prima modalità, quindi partendo da fcc ("config.0") per $10^4$ step (misurando il valore della temperatura una volta ogni 10 steps in modo da ottenere $10^3$ misure). Come si vede dalla figura sottostante, il sistema non sta oscillando in un intorno della temperatura desiderata, perciò si fa ripartire la simulazione nella seconda modalita, e si ripete questo per un totale di 7 volte correggendo le velocità. Viene accettata come ragionevole la configurazione ottenuta dopo 7 riequilibrazioni e viene salvata la configurazione delle particelle, in modo da poter essere recuperata in un secondo momento.

Tutte le quantità ottenute sono state poi ritrasformate in unità del SI.

```python
K_B=1.380649*10**(-23)
a_m_u=1.66053906660*10**(-27)
sigma=0.34*10**(-9)
epsilon=120*K_B
mass=39.948*a_m_u

fs_temp=epsilon/K_B
fs_step=sigma*math.sqrt(mass/epsilon)
fs_dist=sigma
fs_ene=epsilon

temp_rif=0.8

f = np.loadtxt("correction_argon_solid.dat")

x=np.zeros(f.size)
for i in range(f.size):
    x[i]=i+1
x=x*fs_step*10**9
f=f*fs_temp
temp_rif=temp_rif*fs_temp

plt.figure()
plt.plot(x,f, label='measured value')
plt.hlines(temp_rif, 0, x[x.size-1], label='target value')
plt.xlabel('time [ns]')
plt.ylabel('temperature [K]')
plt.grid(True)
plt.title('Solid Argon equilibration')
plt.legend()

plt.show()
```

Viene ora effettuata la simulazione vera e propria nel caso di argon solido. Per iniziare la simulazione da argon solido alla temperatura desiderata è sufficiente importare i parametri di input e la configurazioni finali salvate dopo il punto precedente. La simulazione viene quindi effettuata nella terza modalità, è composta $10^5$ steps e ottiene una misura delle grandezze ogni 10 steps, per un totale di $10^4$ misure.
Le misure effettuate durante la simulazione sono state salvate.
Vengono inoltre salvate anche le medie progressive delle grandezze e i relativi errori all'aumentare del numero di blocchi di misure considerati.

```python
f1, error1 = np.loadtxt("solid_argon_ave_epot.dat", usecols=(0,1), delimiter=' ', unpack='true')
x=np.zeros(f1.size)
for i in range(x.size):
    x[i]=i+1
f1=f1*fs_ene*10**21
error1=error1*fs_ene*10**21

plt.figure()
plt.errorbar(x,f1,error1)
plt.xlabel('#blocks')
plt.ylabel('epot [J x 10^-21]')
plt.grid(True)
plt.title('Solid Argon, measured average potential energy')

plt.figure()
f2, error2 = np.loadtxt("solid_argon_ave_ekin.dat", usecols=(0,1), delimiter=' ', unpack='true')
f2=f2*fs_ene*10**21
error2=error2*fs_ene*10**21
plt.errorbar(x,f2,error2)
plt.xlabel('#blocks')
plt.ylabel('ekin [J x 10^-21]')
plt.grid(True)
plt.title('Solid Argon, measured average kinetic energy')

plt.figure()
f3, error3 = np.loadtxt("solid_argon_ave_etot.dat", usecols=(0,1), delimiter=' ', unpack='true')
f3=f3*fs_ene*10**21
error3=error3*fs_ene*10**21
plt.errorbar(x,f3,error3)
plt.xlabel('#blocks')
plt.ylabel('etot [J x 10^-21]')
plt.grid(True)
plt.title('Solid Argon, measured average total energy')

plt.figure()
f4, error4 = np.loadtxt("solid_argon_ave_temp.dat", usecols=(0,1), delimiter=' ', unpack='true')
f4=f4*fs_temp
error4=error4*fs_temp
plt.errorbar(x,f4,error4, label='measured value')
plt.hlines(temp_rif, 0, x[x.size-1], label='target value')
plt.xlabel('#blocks')
plt.ylabel('temperature [k]')
plt.grid(True)
plt.title('Solid Argon, measured average temperature')
plt.legend()

plt.show()
```
